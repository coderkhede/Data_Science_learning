```python
string1=input("Enter a string: ")
string1=string1.lower()
each_char=tuple(string1)
chars=list(each_char)

dict10={}
for c in chars:
    if c in dict10:
        dict10[c]+=1
    else:
        dict10[c]=1
print(dict10)
```
