### A14

```python
num=int(input("enter a number: "))
for i in range (2,num):
    if num%i==0:
        break
    else:
        print("The number is prime")
```

### A15

```python
num=int(input("enter a number: "))
prvnum=0
fab=1
for i in range(3,num+1):
    print(prvnum+fab)
    temp=fab
    fab=prvnum+fab
    prvnum=temp
print("The Fibonacci series up to", num, "is ",fab)
```
