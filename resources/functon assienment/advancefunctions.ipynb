```python
def key_value_args(**kwargs):
    return {key:value for key, value in kwargs.items() if isinstance(value, int)}
print(key_value_args(name="Alice", age=30, city="New York", score=95))
print(key_value_args(product="Laptop", price=1200, stock=50, brand="Dell"))
print(key_value_args(country="USA", population=331000000, area=9833520, currency="USD"))
```

```python
def time_calculator(func):
    import time
    def wrapper(*args, **kwargs):
        start_time = time.time()
        result = func(*args, **kwargs)
        end_time = time.time()
        print(f"Execution time: {end_time - start_time} seconds")
        return result
    return wrapper
@time_calculator
def compute_sum(n):
    return sum(range(n + 1))
print(compute_sum(1000000))
print(compute_sum(5000000))

```

```python
def high_order_function(filter, map,list):
    filtered_list = filter(list)
    mapped_list = map(filtered_list)
    return mapped_list
print(high_order_function(lambda x: [i for i in x if i % 2 == 0],lambda x: [i**2 for i in x],[1,2,3,4,5,6]))
print(high_order_function(lambda x: [i for i in x if i > 3],lambda x: [i*10 for i in x],[1,2,3,4,5,6]))
```

```python
def avrg(list=[]):
    if list == []:
        return None
    else:
        return sum(list)/len(list)
print(avrg([10,20,30,40,50]))
print(avrg([]))
print(avrg([5,15,25]))
print(avrg([100,200,300]))
print(avrg([]))
```
