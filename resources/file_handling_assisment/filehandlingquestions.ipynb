```python
list1=[]
item="start"
while item!="":
    item=input("Enter a line to add to the document (press Enter to finish): ")
    if item!="":
        list1.append(item +"\n")
with open("document.txt","w") as file:
    file.writelines(list1)
```

```python
import os
sum_of_word=0
with open("document.txt","r") as file:
    for line in file:
        sum_of_word+=len(line.split())
print("Total number of words in document.txt:",sum_of_word)
```
